```python
# 🚀 AI Startup Idea Generator (No API Version)

# This version does NOT require:
# ❌ OpenAI API
# ❌ Billing
# ❌ API keys

# It uses prompt engineering structure + rule-based generation
# to simulate an AI startup idea generator.

import random

startup_names = [
    "NovaLaunch",
    "SkillSpark",
    "EduFlow",
    "HealthSync",
    "FinPilot",
    "AgriBoost",
    "StudySprint",
    "QuickHire",
    "EcoTrack",
    "MindMesh"
]

problems = {
    "EdTech": [
        "Students struggle with last-minute exam preparation",
        "Learners find online courses too long and boring",
        "Students lack personalized study plans"
    ],
    "Health": [
        "People struggle to track daily health habits",
        "Patients miss medication schedules",
        "Mental health support is inaccessible"
    ],
    "FinTech": [
        "Young adults struggle with budgeting",
        "Small businesses lack financial planning tools",
        "People find investing too complicated"
    ]
}

solutions = {
    "EdTech": [
        "AI-generated quick revision guides",
        "Micro-learning video platform",
        "Personalized study planner"
    ],
    "Health": [
        "Habit tracking mobile app",
        "Smart medication reminder system",
        "AI wellness assistant"
    ],
    "FinTech": [
        "Simple budgeting dashboard",
        "AI finance assistant for SMBs",
        "Beginner-friendly investing app"
    ]
}

business_models = [
    "Freemium subscription",
    "Monthly subscription",
    "One-time purchase",
    "Ad-supported model",
    "B2B SaaS pricing"
]

usps = [
    "Highly personalized experience",
    "Simple and beginner-friendly",
    "Affordable for students",
    "AI-powered recommendations",
    "Fast and lightweight platform"
]

risks = [
    "High competition",
    "User retention challenges",
    "Scaling infrastructure costs",
    "Market adoption uncertainty",
    "Data privacy concerns"
]


def generate_startup_idea(industry, audience, budget):
    name = random.choice(startup_names)

    problem = random.choice(
        problems.get(industry, ["People face inefficiencies in daily tasks"])
    )

    solution = random.choice(
        solutions.get(industry, ["AI-powered productivity platform"])
    )

    model = random.choice(business_models)
    usp = random.choice(usps)
    risk = random.choice(risks)

    output = f"""
🚀 Startup Idea Generator

Industry: {industry}
Target Audience: {audience}
Budget: {budget}

----------------------------------------

1. Startup Name:
{name}

2. Problem:
{problem}

3. Solution:
{solution}

4. Target Market:
{audience}

5. Business Model:
{model}

6. Competitive Advantage (USP):
{usp}

7. Potential Risks:
{risk}

----------------------------------------
"""

    return output


# MAIN PROGRAM

print("🚀 Welcome to the AI Startup Idea Generator\n")

industry = input("Enter Industry (EdTech / Health / FinTech): ")
audience = input("Enter Target Audience: ")
budget = input("Enter Budget Level (Low / Medium / High): ")

idea = generate_startup_idea(industry, audience, budget)

print(idea)

# Save output
with open("startup_idea.txt", "w") as file:
    file.write(idea)

print("✅ Startup idea saved to startup_idea.txt")
```
